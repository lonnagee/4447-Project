## Running through the entire dataset through vader


```python
import pandas as pd
import numpy as np
import seaborn as sns
import nltk
import regex
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
reviews_df = pd.read_json('/Users/sean/Documents/University_of_denver/comp_4447_tools_1/Group_project/yelp_dataset/yelp_academic_dataset_review.json',lines =True, nrows = 5000)
```

```python
reviews_df.head()
```

```python
cols_to_drop =["review_id", "user_id", "useful", "funny", "cool"]
reviews_df.drop(cols_to_drop, inplace= True, axis=1)
reviews_df.head()
```

```python
sns.set_palette('rocket')
sns.countplot(x = 'stars', data = reviews_df)
```

```python
reviews_df.head()
```

```python
scores = []
for i in range(len(reviews_df)):
    # get the text in the review
    sentence = reviews_df.iloc[i][2].strip()
    # tokenize the text
    tokens = word_tokenize(sentence.lower())
    tokens = [x for x in tokens if x not in stopwords.words('english') and x.isalpha()]
    # lematize the tokens
    lemma = WordNetLemmatizer()
    tokens = [lemma.lemmatize(x) for x in tokens]
    final_text = ' '.join(tokens)
    # analyze the sentence
    analyzer = SentimentIntensityAnalyzer()
    score = analyzer.polarity_scores(final_text)
    scores.append(score)
```

```python
scores
```

```python
scores_df = pd.DataFrame(scores)
```

```python
new_df = pd.concat([reviews_df, scores_df], axis=1) 
```

```python
min_norm = 1
max_norm = 5
min_score = min(scores_df['compound'])
print(min_score)
max_score = max(scores_df['compound'])
print(max_score)
```

```python
def cmpd_to_stars(compound):
    min_cmpd = np.min(compound)
    max_cmpd = np.max(compound)
    
    b = np.round(((compound - min_cmpd)/(max_cmpd - min_cmpd)*4)+1)
    return b
```

```python
new_df['stars_pred'] = cmpd_to_stars(new_df['compound'])
```
